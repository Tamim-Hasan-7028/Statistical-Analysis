Correlation interpretation does not prove that one variable caused some change in another, but it can suggest there is an association between the two variables. It also generally looks for linear relationships between variables and cannot determine if a third unaccounted for variable might be affecting results.

When r is 0: no relationship.

When r is 1: strong positive relationship, or the same direction of movement.

When r is -1: strong negative relationship, or a reverse direction of movement.

0.1-0.3 is weak, .4-.7 is modarate, .8-1 is strong correlation.

## **Transforming variables**

When variables have skewed distributions, they often require a transformation in order to form a linear relationship with another variable so that correlation can be computed.


---



1) Create a scatterplot of **happiness_score** vs **gdp_per_cap** and calculate the correlation between them.

2) Add a new column to **world_happiness** called **log_gdp_per_cap** that contains the log of **gdp_per_cap**.

  Create a seaborn scatterplot of **log_gdp_per_cap** and **happiness_score** and calculate the correlation between them.

```python
#1
sns.scatterplot(x='happiness_score',y='gdp_per_cap',data=world_happiness)
plt.show()
print(world_happiness['happiness_score'].corr(world_happiness['gdp_per_cap']))
```

So, this is skewed. But still showing correlation. This is an error. We can only apply correlation on linear graph.

But if we use log transformation, then we can convert this graph into a linear graph.

```python
#2 Transformation
world_happiness['log_gdp']=np.log(world_happiness['gdp_per_cap'])
print(world_happiness.head())
```

```python
sns.scatterplot(x='happiness_score',y='log_gdp',data=world_happiness)
plt.show()
```

So, in this way we are able to transform the graph into a linear graph. Now we can find correlation from those variables.

```python
print(world_happiness['happiness_score'].corr(world_happiness['log_gdp']))
```

So, this is a positive strong correlation. So we can say that there is an association between those variables.
